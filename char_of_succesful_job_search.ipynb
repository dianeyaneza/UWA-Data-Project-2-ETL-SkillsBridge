```python
import pandas as pd
import numpy as np
```

```python
# Read excel sheet
file_name = "Resources\Characteristics of successful and unsuccessful job search experience.xls"
df = pd.read_excel(file_name, sheet_name="Table 13.1")
df
```

```python
# Assign row as column headers
df.columns = df.iloc[4]
df_clean = df.copy()
df_clean
```

```python
# Delete first 7 rows
df_clean = df.drop(df.index[0:7]).copy()
df_clean
```

```python
# Reset numbering of index column
df_clean = df_clean.reset_index(drop=True)
# Remove column name of index
df_clean.columns.name = None
df_clean
```

```python
# Copy first 46 rows
df_clean = df_clean.drop(df_clean.index[46:]).copy()
df_clean
```

```python
#Drop NaN columns
df_clean.dropna(how="all", inplace=True, axis=1)
df_clean
```

```python
#Drop rows with any NaN
df_clean.dropna(how="any", inplace=True, axis=0)
df_clean
```

```python
#Rename first column header from 'NaN'
df_clean.columns.values[0] = "Characteristics of successful & unsuccessful job search"
df_clean
```

```python
#Rename columns ot differentiate unemployed persons vs. employed persons who started their current job in the previous 12 months 
df_clean.columns.values[1:] = ["unemployed_males",	"unemployed_females",	 "unemployed_persons",	"employed_males",	"employed_females",	"employed_persons"]
df_clean
```

```python
# Reset numbering of index column
df_clean = df_clean.reset_index(drop=True)
df_clean
```

```python
# Keep state records only
df_clean_by_state = df_clean.drop(df.index[8:42]).copy()
df_clean.columns.values[0] = "state_territory"
df_clean_by_state[["unemployed_males",	"unemployed_females",	 "unemployed_persons",	"employed_males",	"employed_females",	"employed_persons"]] = df_clean_by_state[["unemployed_males",	"unemployed_females",	 "unemployed_persons",	"employed_males",	"employed_females",	"employed_persons"]] * 1000
df_clean_by_state
```

```python
df_clean_by_state
```

```python
# Print to csv
df_clean_by_state.to_csv('employed_vs_unemployed_by_state.csv', index=False)
```

```python
# Keep age records only
df_clean_by_age = df_clean.iloc[8:14].copy()
df_clean.columns.values[0] = "age_range"
df_clean = df_clean.reset_index(drop=True)
df_clean_by_age[["unemployed_males",	"unemployed_females",	 "unemployed_persons",	"employed_males",	"employed_females",	"employed_persons"]] = df_clean_by_age[["unemployed_males",	"unemployed_females",	 "unemployed_persons",	"employed_males",	"employed_females",	"employed_persons"]] * 1000

df_clean_by_age
```

```python
# Print to csv
df_clean_by_age.to_csv('employed_vs_unemployed_by_age.csv', index=False)
```

```python
# Keep Number of employers in previos 12 months records only
df_clean_by_num_pre_employers = df_clean.iloc[14:19].copy()
df_clean.columns.values[0] = "Number of previous employers in last 12 months"
df_clean_by_num_pre_employers
```

```python
# Print to csv
df_clean_by_num_pre_employers.to_csv('employed_vs_unemployed_by_num_prev_empl.csv', index=False)
```

```python
# Keep Steps taken to look for more work records only
df_clean_by_steps_taken = df_clean.iloc[19:31].copy()
df_clean.columns.values[0] = "Steps taken to search for work"
df_clean_by_steps_taken
```

```python
# Print to csv
df_clean_by_steps_taken.to_csv('employed_vs_unemployed_by_steps_taken.csv', index=False)
```

```python
# Keep Steps taken to look for more work records only
df_clean_by_highest_qual = df_clean.iloc[32:42].copy()
df_clean.columns.values[0] = "Highest non-school qualification"
# Remove redundant column
df_clean_by_highest_qual.drop([32])
```

```python
# Print to csv
df_clean_by_highest_qual.to_csv('employed_vs_unemployed_by_quals.csv', index=False)
```
